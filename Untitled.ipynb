```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data_set.csv")
df.head()
```

```python
df.info()
df.describe()
df.columns
```

```python
df = df.loc[:, ~df.columns.duplicated()]
```

```python
df.isnull().sum()
```

```python
df['YEAR'] = df['YEAR'].astype(int)
df['MONTH'] = df['MONTH'].astype(int)
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_").str.lower()
```

```python
df['total_sales'] = df['retail_sales'] + df['warehouse_sales']

df['transfer_ratio'] = df['retail_transfers'] / df['total_sales']

df['demand'] = df['total_sales'] + df['retail_transfers']
```

```python
df['date'] = pd.to_datetime(df[['year', 'month']].assign(day=1))
```

```python
df.fillna(0, inplace=True)
```

```python
df.to_csv("cleaned_data.csv", index=False)
```

```python
df.groupby('date')['total_sales'].sum().plot()
plt.title("Total Sales Trend")
plt.show()
```

```python
df.groupby('item_type')['total_sales'].sum().plot(kind='bar')
plt.title("Sales by Item Type")
plt.show()
```

```python
df.groupby('supplier')['total_sales'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top Suppliers")
plt.show()
```
